```python
def plt_pie_on_ax(ax, sr, decimals=1):
    """
    Dibuja una funcin seno en un objeto Axes de Matplotlib proporcionado.
    
    Args:
        ax (matplotlib.axes.Axes): El eje donde se dibujar el grfico.
    """
    pcts_fmt = f'%.{decimals}f%%'
    ax.pie(x=sr.values, autopct=pcts_fmt)
    ax.legend(sr.index)
```

```python
def plt_sin(ax=None):
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    if ax is None:
        fig, ax = plt.subplots(figsize=(3, 1))
    ax.plot(x, y)
    ax.set_title("Grfica de Seno")
    return ax

def plt_pie(ax=None):
    if ax is None:
        fig, ax = plt.subplots(figsize=(3, 3))
    ax.pie(x=[60, 30, 10], labels=['A', 'B', 'C'],
           startangle=90, autopct='%1.1f%%')
    ax.set_title("Grfica de Torta")
    return ax

# Nueva figura combinada
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4))
plt_sin(ax1)
plt_pie(ax2)
plt.tight_layout()
plt.show()
```

```python
def plot_sin_on_ax(ax):
    """
    Dibuja una funcin seno en un objeto Axes de Matplotlib proporcionado.
    
    Args:
        ax (matplotlib.axes.Axes): El eje donde se dibujar el grfico.
    """
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    ax.plot(x, y)
    ax.set_title("Grfica de Seno")
    ax.set_xlabel("x")
    ax.set_ylabel("sin(x)")
    ax.grid(True, linestyle='--', alpha=0.6)

def plot_pie_on_ax(ax):
    """
    Dibuja un grfico de torta en un objeto Axes de Matplotlib proporcionado.

    Args:
        ax (matplotlib.axes.Axes): El eje donde se dibujar el grfico.
    """
    data = [60, 30, 10]
    labels = ['A', 'B', 'C']
    ax.pie(data, labels=labels, autopct='%1.1f%%', startangle=90)
    ax.set_title("Grfico de Torta")
    # Asegura que el grfico de torta sea un crculo.
    ax.axis('equal')  

# --- Script Principal para Combinar los Grficos ---

# 1. Crear una figura y una grilla de subplots. 
#    1 fila, 2 columnas. figsize ajusta el tamao total de la figura.
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# 2. Aadir un ttulo general a toda la figura.
fig.suptitle('Figura Combinada con Subplots', fontsize=16)

# 3. Llamar a cada funcin para que dibuje en su respectivo subplot.
plot_sin_on_ax(ax1)
plot_pie_on_ax(ax2)

# 4. Ajustar el espaciado entre subplots para que no se superpongan los ttulos.
plt.tight_layout(rect=[0, 0, 1, 0.95]) # El rect ajusta el espacio para el suptitle

# 5. Mostrar la figura final.
plt.show()
```
